## Task 3

```python
def rename_my_file(old_file_name):
    month={
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }
    filemonth=old_file_name.split('-')[0]
    date_country=old_file_name.split('-')[-1]
    for key,value in month.items():
        if key==filemonth:
            new_filename=f'{value}-{date_country}'
    return new_filename

old_file_list=['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
for file in old_file_list:
    new_file_name=rename_my_file(file)
    print(f'{file}---->{new_file_name}')
```

## Task 4

```python
import glob
old_file_path_list=glob.glob('os-collaboration-exercise-data/*.txt',recursive=True)
for file in old_file_path_list:
    old_filename=file.split(os.path.sep)[-1]
    if os.path.exists(file):
        new_filename=rename_my_file(old_filename)
        print(f'{old_filename}---->{new_filename}')
    else:
        print(f'The file has already been renamed!')
```
